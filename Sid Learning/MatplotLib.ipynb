```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y= x**2
```

```python
x
```

```python
y
```

```python
# Functional method
plt.plot(x,y,'r');
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('first plot')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r');
plt.subplot(1,2,2)
plt.plot(y,x,'b');
```

```python
# Object Oriented Method
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('x Label')
axes.set_ylabel('y Label')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y);
axes2.plot(y,x);
axes2.set_title('Smaller plot')
axes1.set_title('larger plot')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
```

```python
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
fig
```

```python
axes[0].plot(x,y)
fig

axes[1].plot(y,x)
fig
```
