Seaborn is a data visualization package. 
Distribution Plots: These help in visualizing the distribution of a data set.
1. distplot
2. jointplot
3. kdeplot
4. pairplot
5. rugplot

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as ml
%matplotlib inline
```

```python
# seaborn comes with built in data sets
tips = sns.load_dataset('tips')
```

```python
type(tips)
tips.info()
tips.head(1)
```

```python
sns.distplot(tips['total_bill'],kde=False, bins = 20)
```

```python
sns.distplot(tips['total_bill'],kde=True, )
```

```python
sns.jointplot(x=tips['total_bill'],y=tips['tip'],data=tips, kind = 'scatter')
```

```python
sns.jointplot(x=tips['total_bill'],y=tips['tip'],data=tips, kind = 'reg')
```

```python
sns.jointplot(x=tips['total_bill'],y=tips['tip'],data=tips, kind = 'hex')
```

```python
#sns.pairplot(data=tips, hue='sex')
#sns.pairplot(data=tips, hue='sex' palette='coolwarm')
sns.pairplot(data=tips, hue='sex', kind ='reg')
```

```python
fig = ml.figure()
axes = fig.add_axes([0,0,1,1])
sns.rugplot(tips['total_bill'])
sns.kdeplot(tips['total_bill'])
```

# Session 2 on Seaborn - Categorical Data Plots

Bar plot and count plot

```python
tip = sns.load_dataset('tips')
import numpy as np
```

```python
sns.barplot(x='sex',y=tips['total_bill'],data=tips)
# when using categorical values on x, bar plot gives an aggrigation on Y. by default its mean total bill for each sex.
```

```python
sns.barplot(x=tips['sex'],y=tips['total_bill'],data=tips,estimator=np.std)
```

```python
sns.countplot(x=tips['sex'],data=tips)
```

```python
fig = ml.figure(figsize=(10,5))
sns.boxplot(x='day',y='total_bill', data=tips, hue='sex')
```

```python
fig = ml.figure(figsize=(10,5))
sns.violinplot(x='day',y='total_bill', data=tips, hue='sex')
```

```python
fig = ml.figure(figsize=(10,5))
sns.violinplot(x='day',y='total_bill', data=tips, hue='sex' , split=True)
```

```python
fig = ml.figure(figsize=(10,5))
sns.swarmplot(x='day',y='total_bill', data=tips, hue='sex')
```

```python
sns.factorplot(x='day',y='total_bill', data=tips, kind='bar')
```

# Session 3 on Grids

```python
iris = sns.load_dataset('iris')
```

```python
g = sns.PairGrid(iris)
g.map(ml.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(ml.hist)
g.map_upper(ml.scatter)
g.map_lower(sns.kdeplot)
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species')
```
